**Plot and obaservations**

```python
# Plot all categorical features
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.reshape(-1)

for idx, col in enumerate(cat_col):
    sns.countplot(x=col, hue='score_miss', data=all_player_shots, palette='Paired', ax=axes[idx])
    axes[idx].set_title('Count Plot by ' + col)
    axes[idx].set_xlabel('')

    # Annotating with percentage
    total_count = len(all_player_shots[col])
    for p in axes[idx].patches:
        height = p.get_height()
        percentage = f"{100 * height / total_count:.1f}%"
        axes[idx].annotate(percentage, (p.get_x() + p.get_width() / 2, height),
                           ha='center', va='bottom')
    axes[idx].legend(loc='upper left', bbox_to_anchor=(1.0, 1.0), ncol=1)
    axes[idx].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

Observations:
- 2PT_PAINT tends to score more than miss
- More shots were done in the first to third quarters
- Home team doesn't have an obvious advantage in this
- G shoots the most with a lower score ratio. F-C doesn't shoot a lot but has the highest score ratio.

The closest_opponent_dist, height, and opponent_height have outliers.

```python
def remove_outliers(df, column_name, threshold=1.5):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    filtered_df = df.loc[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    return filtered_df
```

```python
all_player_shots_2 = remove_outliers(all_player_shots, 'closest_opponent_dist')
```

```python
# Plot all num_col against shotType

for num in num_col:
    plt.figure(figsize=(9, 3))
    ax = sns.violinplot(y=num, x ='shotType', hue='score_miss', data = all_player_shots_2, palette='RdBu', dodge=False,
                       linewidth=0.5, cut = 0)
    ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0), ncol=1)
    sns.despine(left=True)
    plt.setp(ax.collections, alpha=0.7)
    plt.show()
```

Observations:
- 2PT_SHOT is more likely to score at the end of a quarter
- 3PT_DRIBBLE SHOT and 2PT_DRIBBLE SHOT are more likely to miss when the shotClock starts
- 2PT_SHOT has a good score chance at shotClock [10,15] and when shot when y is small (facing the basket)
- 2PT_PAINT and 2PT_DRIBBLE SHOT have a higher chance to miss when an opponent is close
- 2PT_PAINT has a higher chance to score when the shooter is tall
- 3PT_DRIBBLE SHOT has a higher chance to score when the closest opponent is short

```python
scored = all_player_shots[all_player_shots['score_miss'] == 'score']
missed = all_player_shots[all_player_shots['score_miss'] == 'miss']
```

```python
for num in num_col:
    fig, ax = plt.subplots(figsize=(6,4))
    sns.distplot(scored[num], hist=True, kde=True, 
                 kde_kws={'linewidth':1, 'color': 'darkblue'}, bins=20, color='lightblue', label='score')
    sns.distplot(missed[num], hist=True, kde=True, 
                 kde_kws={'linewidth':1, 'color': 'goldenrod'}, bins=20, color='tan', label='miss')
    plt.legend()
    plt.show()
```

Observations:
- The shots are more likely to score when shotClock > 10
- The shots are more likely to miss at mid-range
- The shots are more likely to score when y is small (facing the basket)
- The shots are more likely to score when shooter is taller then 80: (7'-8")
- The shots are more likely to miss when closest opponent importance is > 0.04
- The shots are more likely to miss when closest opponent is taller then 79" (7'-7")
